```python
#dropna()
df = pd.read_csv(path)
df
```

```python
df.dropna() #drops rows with null values
```

```python
#axis = 0 - drop row
#axis = 1 - drop col
df.dropna(axis=1)
```

```python
df.dropna(how = "any")
```

```python
df.dropna(how = "all") #all values in row are null, then it drops that row
```

```python
#thresh - drops min NaN
df.dropna(thresh=1)
```

```python
df.dropna(axis=0,thresh=5)
```

```python
df.dropna(subset=["MOVIE_TITLE","RATING"])
```

```python
#inplace - drops all NaN rows
df.dropna(inplace=True) #changes original df
df
```

```python
df.dropna(inplace=False)
```

Join can join 2 DataFrames.
Cannot join name column name, therefore, use lsuffix or rsuffix

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({1:["A"],2:"B",3:"C"})
df2 = pd.DataFrame({4:["Bubble"],5:"Ham",6:"Bee"})
```

```python
display(df1,df2)
```

```python
df1.join(df2) #requires different key values
```

```python
df2.join(df1)
```

```python
display(df1,df2)
```

```python
#unmatched index
df1 = pd.DataFrame({1:["A"],2:"B",3:"C"}, index=["a"])
df2 = pd.DataFrame({4:["Bubble"],5:"Ham",6:"Bee"})
```

```python
display(df1,df2) #different indexes
```

```python
df1.join(df2)
```

```python
#unmatched elements
df1 = pd.DataFrame({1:["A","Apple"],2:["B","Ball"],3:["C","Cat"]})
df2 = pd.DataFrame({4:["Bubble"],5:"Ham",6:"Bee"})
```

```python
display(df1,df2) #2 rows in df1, 1 row in df2
```

```python
df1.join(df2) 
```

```python
df1 = pd.DataFrame({1:["A"],2:"B",3:"C"})
df2 = pd.DataFrame({1:["Bubble"],5:"Ham",6:"Bee"})
df3 = pd.DataFrame({1:[10],2:459,3:83})
display(df1,df2,df3)
```

```python
df1.join(df2,lsuffix="_a")
```

```python
df1.join(df2,rsuffix="_b")
```

Append

```python
d1 = pd.DataFrame({"ID":[1,2,3,4,5],"Class":[10,20,30,40,50]})
d2 = pd.DataFrame({"ID":[1,2,3,4,5],"Class":[11,22,33,44,55]})
display(d1,d2)
```

```python
d1.append(d2,ignore_index=True)
```

```python
d1.append(d2,ignore_index=True,sort=True)
```

```python
d1 = pd.DataFrame({"ID":[1,2,3],"Class":[10,20,30]})
d2 = pd.DataFrame({"ID":[1,2,3,4,5],"Class":[11,22,33,44,55]})
display(df1,df2)
```

```python
d1.append(d2,sort=True)
```

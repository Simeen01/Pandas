Filling NA values

```python
import pandas as pd
```

```python
path = "/content/drive/MyDrive/Datasets/Movieratingdata2.csv"
df = pd.read_csv(path)
```

```python
df
```

```python
#Replacing NaN with 1
df.fillna(1) #scalar
```

```python
df.fillna({"MOVIE_ID":2,"PRODUCTION_YEAR":3})
```

```python
#pad and ffill give same output
df.fillna(method="ffill") #forward fill
```

```python
#backfill and bfill give same output
df.fillna(method="bfill") #backward fill
```

```python
df.fillna(method="pad")
```

```python
df.fillna(method="ffill",axis=0) #row
```

```python
df.fillna(method="ffill",axis = 1) #column
```

```python
df.fillna(5,axis=0)
```

```python
df.fillna(method="ffill",limit = 2)
```

```python
df.fillna(method="bfill",inplace=True)
```

```python
df
```

```python
df.fillna(method="bfill",inplace=False)
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({"Id":[1,2,3,4,5],"Class":[10,20,30,40,50]})
df1
```

```python
df2 = pd.DataFrame({"Id":[1,2,3,4,5],"Class":['A','B','C','D','E']})
df2
```

```python
#on - common column
pd.merge(df1,df2,on="Id")
```

```python
#suffixes - naming same column with suffix each char to consecutive column
pd.merge(df1,df2,on="Id",suffixes="12")
```

```python
df3 = pd.DataFrame({"Id":[1,2,3,4,5],"Name":["Simeen","Saaliha","Safaa","Ayesha","Alia"]})
df3
```

```python
pd.merge(df1,df3,on="Id")
```

```python
pd.merge(df1,df2,df3,on="Id")
```

```python
pd.merge(df1,df3,how="right")
```

```python
d = pd.DataFrame({"Id":[1,2,3],"Blaah":[34,65,276]})
d
```

```python
pd.merge(df1,d,on="Id",how="inner")
```

```python
pd.merge(df1,d,on="Id",how="left")
```

```python
pd.merge(df1,d,on="Id",how="right")
```

```python
#displays all
pd.merge(df1,d,on="Id",how="outer")
```

```python
#indicator - shows merging style
pd.merge(df1,d,on="Id",how="outer",indicator=True)
```

```python
t = pd.DataFrame({"Id":[1,2,3,4],"Whatever":["Ted","Raff",45,"GOAT"]})
```

```python
t
```

```python
pd.merge(df1,t,left_index=True,right_index=True)
```

```python
r = pd.DataFrame({"Id":[1,2,3,5,6],"Class":[10,20,30,40,50]})

y = pd.DataFrame({"Id":[1,2,3,4,7],"Whatever":["Ted","Raff",45,"GOAT","Zap"]})
```

```python
r
```

```python
y
```

```python
pd.merge(r,y,left_index=True,right_index=True)
```

```python
#all values of both tables merged
pd.merge(r,y,on="Id",how="outer")
```

```python
#left table's values - r
pd.merge(r,y,on="Id",how="left")
```

```python
#right table's values - y
pd.merge(r,y,on="Id",how="right")
```

```python
#common Id values
pd.merge(r,y,on="Id",how="inner")
```

```python
a = pd.Series([1,2,3,4,5])
a
```

```python
b = pd.Series([6,7,8,9,10])
b
```

```python
c = pd.Series(["A","e","i","o","u"],name="c")
c
```

```python
pd.concat([a,b],axis=0)
```

```python
pd.concat([a,b,c],axis=0) #axis = 0 //rowise
```

```python
#ignore_index= True //ignore's other tables index and puts sequential index
pd.concat([a,b,c],axis=0,ignore_index=True)
```

```python
pd.concat([a,b,c],axis=1) #axis = 1 //column-wise
```

```python
#Keys - table merge as 1 row
pd.concat([a,b,c],axis=0,keys=["A","B","C"])
```

```python
pd.concat([a,b,c],axis=1,keys=[1,2,3])
```
